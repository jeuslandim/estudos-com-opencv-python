Detector de canto Shi-Tomasi e bons recursos para rastrear
link: https://docs.opencv.org/4.x/d4/d8c/tutorial_py_shi_tomasi.html


```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

# Caregue a imagem e a converter para a escala de cinza
cases = {
    1: 'imagens/baby_2.jpg',
    2: 'imagens/baby_1.jpg',
    3: 'imagens/baby_5.jpg',
    4: 'imagens/shitomasi_block1.jpg'
}
img = cases[4] # Path da imagem

img = cv.imread(img)
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
 
corners = cv.goodFeaturesToTrack(gray,10,0.1,20)
corners = corners.astype(np.int64)
 
for i in corners:
    x,y = i.ravel()
    cv.circle(img, (x, y), 3, (255, 0, 0), -1)  # Azul para destacar
 
# Converter BGR para RGB antes de exibir no Matplotlib
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.axis("off")  # Esconde os eixos para uma melhor visualização
plt.show()
```
